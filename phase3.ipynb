## Advance Operation With Data

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021,2022,2023,2024]

sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
    ])
```

```python
print("Sales data shape", sales_data.shape)
```

```python
print("Sample data for 1st 3 restau:\n ", sales_data[:, 1:]) 

""" "[:3]" -> select first 3 elements in the list, 
"[:]" -> select all elements in the list, "[-3:]" -> select until last elements in the list 
"[:, 1:]" -> select all rows and 2nd column onwards in the 2D list """
```

```python
# total sales per year 
print("Total sales per year:",np.sum(sales_data[:, 1:], axis=0)) 

""" axis -> 0 means sum along the columns, 1 means sum along the rows """

# Minimum sales per restau:
print("\nMin sales per restau:",np.min(sales_data[:, 1:], axis=1))
```

```python
 # Maximum sales per year
print("Max sales per year: ", np.max(sales_data[:,1:], axis=0))

# Avarage sales per restau

print("\nAverage sales per restau:", np.mean(sales_data[:,1:], axis=1))
```

```python
# Cumlative sales
cumsum=np.cumsum(sales_data[:,1:],axis=1)
print("cumlative sales:\n",cumsum)

plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumlative sales accross all restaurant")
plt.xlabel("Time")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```

```python
#Vector addition
vector1 = np.array([1,2,3,4])
vector2 = np.array([5,6,7,8])

# print("vector additon of two vectors: ", np.add(vector1, vector2))
print("vector additon of two vectors: ", vector1 + vector2)
# print("vector multiplication:" , np.multiply(vector1,vector2))
print("vector multiplication:" , vector1*vector2)

print("\n Dot product of two vectors: ", np.dot(vector1,vector2))
print("\n Magnitude of vector1: ", np.linalg.norm(vector1))
print("\n Magnitude of vector2: ", np.linalg.norm(vector2))
print("\n Angle b/w two vectors: ", np.arccos(np.dot(vector1,vector2)/((np.linalg.norm(vector1))*(np.linalg.norm(vector2)))))

```
